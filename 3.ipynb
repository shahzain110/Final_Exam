**Maximum Temperature for Dataset**

```python
max_temp_index = data['Temperature'].idxmax()

# Get the row with the maximum temperature
max_temp_row = data.loc[max_temp_index]

# Display the result
print("The maximum temperature in data is  > ", max_temp_row['Temperature'])
```

**Creating new column and categorising data as windy and non-windy**

```python
new_column = []
for wind_speed in data['Wind Speed']:
  # print(wind_speed)
  if wind_speed > 30:
    new_column.append('Windy')
  else:
    new_column.append('Non-Windy')

data['Windy/Non-Windy'] = new_column
```
