```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Final Exam/colabs/dataset.csv')

month_data = data[:31]
month_data
```

```python
data = month_data
# 1. Bar Plot for Daily Precipitation Levels
plt.figure(figsize=(10, 5))
plt.bar(data['Date'], data['Precipitation'], color='skyblue')
plt.xlabel('Date')
plt.ylabel('Precipitation (mm)')
plt.title('Daily Precipitation Levels')
plt.xticks(rotation=45)
plt.show()

# 2. Pie Chart for Precipitation Proportions
# Classify days based on precipitation levels
conditions = [
    (data['Precipitation'] == 0),
    (data['Precipitation'] < 5) & (data['Precipitation'] > 0),
    (data['Precipitation'] >= 5)
]
labels = ['No rain', 'Light rain', 'Heavy rain']
data['Precipitation Category'] = pd.cut(data['Precipitation'], bins=[-0.1, 0, 5, float('inf')], labels=labels)

# Calculate the proportions
proportions = data['Precipitation Category'].value_counts()
```

```python
# Plot the pie chart
plt.figure(figsize=(6, 6))
plt.pie(proportions, labels=proportions.index, autopct='%1.1f%%', startangle=140, colors=['lightgrey', 'lightblue', 'dodgerblue'])
plt.title('Proportion of Days by Precipitation Level')
plt.show()
```
